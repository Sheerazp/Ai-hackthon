```python
#=========____TASK_NO_7____=========
from datetime import datetime

class ParkingFullError(Exception):
    pass

class Vehicle:
    def __init__(self, vehicle_id):
        self.vehicle_id = vehicle_id
        self.entry_time = datetime.now()

class Car(Vehicle):
    pass

class Bike(Vehicle):
    pass

class ParkingLot:
    def __init__(self, capacity=5):
        self.capacity = capacity
        self.slots = {i: None for i in range(1, capacity + 1)}

    def park_vehicle(self, vehicle):
        for slot in self.slots:
            if self.slots[slot] is None:
                self.slots[slot] = vehicle
                print(f"Vehicle {vehicle.vehicle_id} parked at slot {slot}")
                return slot
        raise ParkingFullError("Parking Lot is Full")

    def remove_vehicle(self, vehicle_id):
        for slot, vehicle in self.slots.items():
            if vehicle and vehicle.vehicle_id == vehicle_id:
                hours = (datetime.now() - vehicle.entry_time).seconds / 3600
                fee = max(1, int(hours)) * 2
                self.slots[slot] = None
                print(f"Vehicle {vehicle_id} removed from slot {slot}")
                print(f"Parking fee: ${fee}")
                return fee
        print("Vehicle not found")

    def display_status(self):
        print("\n--- Parking Lot Status ---")
        for slot, vehicle in self.slots.items():
            status = "Empty" if vehicle is None else vehicle.vehicle_id
            print(f"Slot {slot}: {status}")

if __name__ == "__main__":
    try:
        parking_lot = ParkingLot(5)

        car1 = Car("CAR-101")
        bike1 = Bike("BIKE-202")
        car2 = Car("CAR-303")

        parking_lot.park_vehicle(car1)
        parking_lot.park_vehicle(bike1)
        parking_lot.park_vehicle(car2)

        parking_lot.display_status()

        parking_lot.remove_vehicle("CAR-101")
        parking_lot.display_status()

    except ParkingFullError as e:
        print(e)
```
