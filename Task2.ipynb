```python
class Driver:
    def __init__(self, name, location, is_available=True):
        self.name = name
        self.location = location
        self.is_available = is_available

class Rider:
    def __init__(self, location):
        self.location = location

class RideMatcher:
    def __init__(self, drivers):
        self.drivers = drivers

    def find_driver(self, rider):
        nearest_driver = None
        min_distance = float("inf")

        for driver in self.drivers:
            if driver.is_available:
                distance = abs(driver.location - rider.location)
                if distance < min_distance:
                    min_distance = distance
                    nearest_driver = driver

        if nearest_driver:
            nearest_driver.is_available = False
            print(f"Driver Matched: {nearest_driver.name}")
            print(f"Driver Location: {nearest_driver.location}")
            print(f"Rider Location: {rider.location}")
            print(f"Distance: {min_distance}")
            return nearest_driver

        print("No Available Driver found")
        return None

if __name__ == "__main__":
    drivers = [
        Driver("Ali", 2),
        Driver("Ahmed", 6),
        Driver("Sara", 5, is_available=False),
        Driver("Usman", 9),
        Driver("Zara", 4),
    ]

    rider = Rider(location=3)
    matcher = RideMatcher(drivers)
    matcher.find_driver(rider)
```
